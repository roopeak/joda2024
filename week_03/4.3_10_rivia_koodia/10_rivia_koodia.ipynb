# Ryömijät ja raapijat datan keräämisessä

Tässä työkirjassa esitellään yksi tapa kerätä tietoa web-sivuilta käyttämällä Requests- ja BeautifulSoup-kirjastoja.

#### Asennetaan tarvittavat kirjastot

#### Kerätään data

```python
import requests
from bs4 import BeautifulSoup

# Replace URL with the URL of your choice
url = 'https://fi.wikipedia.org/wiki/Tampereen_yliopisto'
response = requests.get(url)

if response.status_code == 200:
	html_content = response.text
	soup = BeautifulSoup(html_content, 'html.parser')
	content = soup.find('body')
	if content:
		data = content.text
		print(data)
	else:
		print('Sisältöä ei löytynyt')
else:
	print('Pyyntö epäonnistui')
```
